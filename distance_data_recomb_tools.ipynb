```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
# reading RAT tool results on similarity plot\

dist_data = pd.read_csv("rat_simplot.csv")
```

```python
dist_data.head()
```

```python
# renamgin df columns, mostly because the position col has blank at the end

dist_data.columns = [ "position", "AF409138.1", "KY829023.3", "Unnamed" ]
```

```python
dist_data.head()
```

```python
dist_data.shape
```

```python
ind = dist_data["position"]
dist_data.index = ind
```

```python
dist_data.head()
```

```python
fig_dist = plt.figure(figsize=(20, 8))
plt.plot(dist_data.loc[:, "AF409138.1"])
plt.plot(dist_data.loc[:, "KY829023.3"])

plt.ylim(0.85, 1.1)

plt.show()
```

```python
fig_dist1 = plt.figure(figsize=(20, 8))
plt.plot(dist_data.loc[:12500, "AF409138.1"], lw=7, alpha=0.7, label="AF409138.1")
plt.plot(dist_data.loc[:12500, "KY829023.3"], lw=7, alpha=0.7, label="KY829023.3")

plt.ylim(0.85, 1.05)
plt.title("distance plot", fontsize=20)
plt.title("distance plot", fontsize=20)
plt.ylabel("distance to saratov lsdv", fontsize=15)

#plt.annotate('recombination event', xy=(6000, 0.98), xytext=(6000, 1.025), fontsize=14, arrowprops=dict(facecolor="grey"))

plt.axvline(5500, alpha=0.5, color="red", lw=3,
            linestyle="dashed", label="putative recombination event break points")
plt.axvline(6500, alpha=0.5, color="red", lw=3,
            linestyle="dashed"  )

plt.legend(prop={"size":20})

plt.show()

fig_dist1.savefig("1st_recomb_event.jpg", fmt="jpg")
```

```python
fig_dist2 = plt.figure(figsize=(20, 8))
plt.plot(dist_data.loc[130000:, "AF409138.1"], lw=7, alpha=0.7, label="AF409138.1")
plt.plot(dist_data.loc[130000:, "KY829023.3"], lw=7, alpha=0.7, label="KY829023.3")

plt.ylim(0.85, 1.05)
plt.title("distance plot", fontsize=20)
plt.ylabel("distance to saratov lsdv", fontsize=15)
plt.xlabel("position", fontsize=15)

plt.axvline(136000, alpha=0.5, color="red", lw=3,
            linestyle="dashed", label="putative recombination event break points")
plt.axvline(138200, alpha=0.5, color="red", lw=3,
            linestyle="dashed")

plt.legend(prop={"size":20})
plt.show()


fig_dist2.savefig("2nd_recomb_event.jpg", fmt="jpg")
```
